# TF-IDF Explained (Term Frequency‚ÄìInverse Document Frequency)

TF-IDF is a statistical measure used in Natural Language Processing (NLP) to evaluate how important a word is to a document in a collection or corpus.

General formula

# tf-idf(t,d) = TF(t,d) * IDF(t)

where 
1) t = term(word)
2) d = document
3) TF(t,f) = term frequency of term t in document d
4) IDF(t) = inverse document frequency of term t across all documents




formula for TF (how frequently a term appears in document d)
 # TF(t,d) = Number of times t appears in d / total terms in d




formula for IDF(t) = measure how import term is across entire corpus
# IDF(t) = log(N / 1 + df(t))
where
1) N = total number of documents
2) df(ùë°) =  number of documents containing term 
3) Adding 1 in denominator prevents division by zero.

```python
from sklearn.feature_extraction.text import TfidfVectorizer


corpus = [
    "Thor eating pizza, Loki is eating pizza, Ironman ate pizza already",
    "Apple is announcing new iphone tomorrow",
    "Tesla is announcing new model-3 tomorrow",
    "Google is announcing new pixel-6 tomorrow",
    "Microsoft is announcing new surface tomorrow",
    "Amazon is announcing new eco-dot tomorrow",
    "I am eating biryani and you are eating grapes",
    "something is amazing",
]
```

```python
v = TfidfVectorizer()
transformed_output = v.fit_transform(corpus)
print(v.vocabulary_)
```

```python
v.get_feature_names_out()
# alphabetically
```

```python
all_feature_names = v.get_feature_names_out()
for word in all_feature_names:
    indx = v.vocabulary_.get(word)
    print(f"{word} || {v.idf_[indx]}")
```

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\hp\OneDrive - National University of Sciences & Technology\Desktop\Ecommerce_data.csv")
df.head()
```

```python
df.label.value_counts()
```

```python
# making new column
df['label_num'] = df.label.map({
    'Household':0,
    'Books':1,
    'Electronics':2,
    'Clothing & Accessories':3 
})
```

```python
df.head(5)
```

# using model on original text

```python
# training model on original text
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df.Text,
    df.label_num,
    test_size = 0.2,
    random_state = 2022,
    stratify = df.label_num
)
```

```python
print("Shape of X_train: ", X_train.shape)
print("Shape of X_test: ", X_test.shape)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report

clf = Pipeline([
    ('vectorizer_tfidf', TfidfVectorizer()),
    ('KNN', KNeighborsClassifier())
])

clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.naive_bayes import MultinomialNB


clf = Pipeline([
    ('vectorizer_tfidf', TfidfVectorizer()),
    ('Multi NB', MultinomialNB())
])

clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier


clf = Pipeline([
    ('vectorizer_tfidf', TfidfVectorizer()),
    ('Random Forest', RandomForestClassifier())
])

clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')

def preprocess(text):
    doc = nlp(text)
    filtered_tokens = []
    for token in doc:
        if token.is_stop or token.is_punct:
            continue
        filtered_tokens.append(str(token.lemma))
    return " ".join(filtered_tokens)
```

```python
df['preprocessed_txt'] = df['Text'].apply(preprocess)
```

# using model on preprocessed text

```python
df.head(5)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df.preprocessed_txt,
    df.label_num,
    test_size = 0.2,
    random_state = 2022,
    stratify = df.label_num
)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report

clf = Pipeline([
    ('vectorizer_tfidf', TfidfVectorizer()),
    ('Random Forest', RandomForestClassifier())
])

clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```
